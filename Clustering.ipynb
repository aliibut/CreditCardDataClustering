```python
from sklearn.preprocessing import StandardScaler, normalize
```

```python
scaler = StandardScaler()
scaled_ccData = scaler.fit_transform(ccData)
norm_ccData = normalize(scaled_ccData)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,1,figsize=(20,12))
sns.heatmap(ccData.corr(),annot=True,linewidths=1)
plt.show()
```

PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(norm_ccData)
```

```python
import numpy as np
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
n_components=2
pca_final = PCA(n_components=n_components)
pca_final.fit(norm_ccData)
pca_ccData = pca_final.fit_transform(norm_ccData)
```

```python
plt.scatter(pca_ccData[:,0], pca_ccData[:,1])
```

```python
for i in np.arange(n_components):
    index =  np.argmax(np.absolute(pca_final.get_covariance()[i]))
    max_cov = pca_final.get_covariance()[i][index]
    column = ccData.columns[index]
    print("Principal Component", i+1, "maximum covariance :", "{:.2f}".format(max_cov), "from column", column)
```

```python
# Interpret PCA
```

```python
from sklearn.cluster import KMeans

sse = {}
n_clust = np.arange(2,11)

for i in n_clust:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(pca_ccData)
    sse[i] = kmeans.inertia_
```

```python
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of Clusters")
plt.ylabel("Within Cluster Sum-of-Squares")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(pca_ccData)
```

```python
plt.scatter(pca_ccData[:,0], pca_ccData[:,1],
            c = KMeans(n_clusters = 3).fit_predict(pca_ccData),
            cmap = plt.cm.summer)
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.show()
```

```python
h = .01

x_min, x_max = pca_ccData[:,0].min() - 1, pca_ccData[:,0].max() + 1
y_min, y_max = pca_ccData[:,1].min() - 1, pca_ccData[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = kmeans.predict(np.array(list(zip(xx.ravel(), yy.ravel()))))

Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.summer,
           aspect='auto', origin='lower')

plt.plot(pca_ccData[:,0], pca_ccData[:,1], 'k.', markersize=2)

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='o', s=10, linewidths=3,
            color='w', zorder=10)

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.xlabel("PC1")
plt.ylabel("PC2")

plt.show()
```

```python
for i in np.arange(len(centroids)):
    print("Center of Cluster", i+1, ":", centroids[i])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Generate random data
np.random.seed(0)
X = np.random.randn(100, 2) * 0.3  # 100 data points with 2 features

# Introduce some random outliers
outliers = np.random.uniform(low=-2, high=2, size=(10, 2))
X = np.vstack([X, outliers])

# Instantiate DBSCAN
epsilon = 0.4  # Maximum distance between two samples for one to be considered as in the neighborhood of the other
min_samples = 5  # The number of samples (or total weight) in a neighborhood for a point to be considered as a core point
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)

# Fit and predict clusters
clusters = dbscan.fit_predict(X)

# Visualize the clustering result
plt.figure(figsize=(10, 6))

# Plot core points (clusters) and noise points (outliers) with different colors
core_points_mask = np.zeros_like(clusters, dtype=bool)
core_points_mask[dbscan.core_sample_indices_] = True

# Plot core points (clusters)
plt.scatter(X[core_points_mask][:, 0], X[core_points_mask][:, 1], c=clusters[core_points_mask], cmap='viridis', marker='o', label='Core Points (Clusters)')

# Plot noise points (outliers)
plt.scatter(X[~core_points_mask][:, 0], X[~core_points_mask][:, 1], c='black', marker='x', label='Noise Points (Outliers)')

plt.title('DBSCAN Clustering Result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
